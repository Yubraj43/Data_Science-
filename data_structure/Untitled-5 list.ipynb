```python
#lagest numbers in a list 
def largest_number(numbers):
    numbers=list(map(int,numbers.split()))
    largest_number=numbers[0]
    for i in numbers:
        if i>largest_number:
            largest_number=i
    return largest_number

numbers=input("Enter numbers separated by space: ")
final=largest_number(numbers)
print("The largest number is:",final)
```

```python
#count element in a list
def count_elements(numbers):
    
    count=0
    for i in numbers:
        count=count+1
    return count

numbers=input("Enter numbers separated by space: ").split()
final=count_elements(numbers)
print("The count of elements is:",final)
```

```python
#reverse a list
def reverse_list(lst):
    lsit=[]
    for i in range(len(lst)-1,-1,-1):
        lsit.append(lst[i])
    return lsit

numbers=input('enter numbers separated by space: ').split(  )
reversed_list=reverse_list(numbers)
print("Reversed list is:",reversed_list)
```

```python
#find factorial of a number
def factorial(n):
    num=1
    for i in range(1,n+1):
        num=num*i   
    return num
number=int(input("Enter a number: "))
fact=factorial(number)  
print("Factorial of",number,"is",fact)
```

```python
#find second largest number in a list



def second_largest(numbers):
    numbers=list(map(int,numbers.split()))
    first= second=float('-inf')

    for i in numbers:
        if i > first:
            second = first
            first = i
        elif i>second and i!=first:
            second=i
    return second
num=input("please enter your number")
result = second_largest(num)
print(result)


```

```python
#merge two list
def merge_lists(list1, list2):
    
    merge_=[]
    for i in list1:
        merge_.append(i)
    for j in list2:
        merge_.append(j)    
    return merge_
list1=input("Enter first list elements separated by space: ").split()
list2=input("Enter second list elements separated by space: ").split()  
merged_list=merge_lists(list1,list2)
print("Merged list is:",merged_list)    
```
