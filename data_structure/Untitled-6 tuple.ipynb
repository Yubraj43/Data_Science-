```python
# sort a tuple without using sort()

# Note: tuples don't have .sort() (that's a list method).

# We can use sorted(), which returns a NEW list, then convert to tuple.



t = (12, 3, 4, 5, 5, 6, 6)



sorted_tuple = tuple(sorted(t))

print('Original tuple:', t)

print('Sorted tuple  :', sorted_tuple)
```

```python
# Find the second largest number in a tuple

t = (5, 3, 4, 2, 1, 6, 7, 8)



# Method 1: Using a loop (finds 2nd largest DISTINCT value)

# We start with very small numbers so any real number is bigger.

first = float('-inf')   # largest so far

second = float('-inf')  # second largest so far



for num in t:

    # If we found a new largest, the old largest becomes second.

    if num > first:

        second = first

        first = num

    # If num is between first and second, update second.

    elif first > num > second:

        second = num



if second == float('-inf'):

    print('No second-largest (need at least 2 distinct values).')

else:

    print('Second largest number is:', second)



# Method 2: Without a loop (sorting)

distinct_sorted = sorted(set(t))

if len(distinct_sorted) < 2:

    print('No second-largest (need at least 2 distinct values).')

else:

    print('Second largest (no-loop) is:', distinct_sorted[-2])
```
