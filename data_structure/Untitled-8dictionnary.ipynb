```python
#Create a dictionary student with keys: name, age, course. Print it.
new={"student":{"name":"Mahat","age":22,"course":"Btech"}}
print(new["student"])
```

```python
#Access a value by key: from student, print the value of name.
print(new["student"]["name"])
```

```python
#What happens if you access a missing key like student["marks"]? Try it.
#print(new["student"]["marks"])
print(new["student"].get("marks"))  # This will print None instead of raising an error
```

```python
#Use get() to safely read a missing key: student.get("marks", 0). What does it return?
print(new["student"].get("marks", 0))  
```

```python
#Add a new key-value pair: add marks: 95 to student.
new["student"]["marks"]=95
print(new["student"])
```
