```python
#Write a function that takes a number n and returns True if it is even, otherwise False
def is_even(n):
    if n% 0 ==0:
        return True
    else:
        return False
```

```python
#Write a function that takes a string and returns the number of vowels in it.
def count_vowels(s):
    vowels='aeiouAEIOU'
    count =0
    for char in s:
        if char in vowels:
            count +=1
    return count
```

```python
#Write a function that takes a number n and returns its factorial.
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
```

```python
#Write a function that takes a list of numbers and returns a new list with only even numbers
def check_even(numbers):
    new =[]
    for i in numbers:
        if i %2 ==0:
            new.append(i)
    return new
```

```python
#Write a function that takes a list of numbers and returns a new list with each element squared
def square_list(numbers):
    new_square=[]
    for i in numbers:
        new_square.append(i*i)
    return new_square
```
