```python
#simple calculator
while True:
    print("Enter operation (+, -, *, /) or 'q' to quit:")
    choice = input( )

    if choice == 'q':
        print()
        break
    if choice in ('+', '-', '*', '/'):
        a=float(input("Enter first number: "))
        b=float(input("enter second number: "))
        if choice =='+':
            print(a+b)
        elif choice =='-':
            print(a - b)
        elif choice =='*':
            print(a*b)
        elif choice =='/':
            if b != 0:
                print(a / b)
            else:
                print("Error: Division by zero is not allowed.")
```

```python
#number guessing game 
import random
number_to_guess = random.randint(1, 100)
attempts = 0
print("Welcome to the Number Guessing Game!")

while True:
    guess =int(input("Enter your guess (1-100) or 0 to quit: "))
    attempts+=1
    if guess < number_to_guess:
        print("Too low! Try again.")
    elif guess > number_to_guess:
        print("Too high! Try again.")
    else:
        print(f"Congratulations! You guessed the number in {attempts} attempts.")
        break
```

```python
#password retry attempt
password = "secure123"
attempts = 3
while attempts > 0:
    user = input("enter your passwod:   ")
    if user == password:
        print("Access granted.")
        break
    else:
        attempts -= 1
        print ("Incorrect password. Try again.", attempts, "attempts left.")
if attempts == 0:
    print("Too many incorrect attempts. Access denied.")
```

```python
#random password generator
import random
import string
length = int(input("Enter desired password length: "))
password =" "
i=0
while i< length:
    password += random.choice(string.ascii_letters + string.digits + string.punctuation)
    i += 1
print("Generated password:", password)
```

```python
blance = 1000
print("please enter your atm pin to proceed :")
pin = input("Enter PIN: ")
if pin == "1234":
        print("PIN accepted.")
        
        while True:
            print("Select transaction: 1. Balance Inquiry 2. Withdraw 3. Deposit 4. Exit")
            choice = input("Enter choice (1-4): ")
            if choice == "1":
                print(f"Your current balance is: ${blance}")
                break
            elif choice == "2":
                amount = float(input("Enter amount to withdraw: "))
                if amount <= blance:
                    blance -= amount
                    print(f"${amount} withdrawn. New balance: ${blance}")
                else:
                    print("Insufficient funds.")
            elif choice == "3":
                amount = float(input("Enter amount to deposit: "))
                blance += amount
                print(f"${amount} deposited. New balance: ${blance}")
            elif choice == "4":
                print("Exiting. Thank you for using the ATM.")
                break
            else:
                print("Invalid choice. Please try again.")
else :
        print("Invalid PIN. Access denied.")
```

```python
#library management system
books = {}
while True:
    print("\n1.Add Book 2.View Books 3.Exit")
    choice = input("enter your choice:")
    if choice == "1":
        title = input( "please enter book title:")
        writer = input("please enter book writer:")
        books[title] = writer
        print(f"Book '{title}' by {writer} added.")
    elif choice == "2":
        print("Books in Library:")
        for title, writer in books.items():
            print(f"'{title}' by {writer}") 
    elif choice == "3":
        print("Exiting library management system.")
        break   
    
```
