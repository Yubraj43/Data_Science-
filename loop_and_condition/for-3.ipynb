```python
#temp to celsius converter
while True:
    c = input("Enter temperature in Celsius (or type 'exit' to quit): ")
    if c == "exit":
        break
    c = float(c)
    f = (c * 9/5) + 32  
    print(f"{c}°C is equal to {f}°F")
```

```python
#prime number checker
num = int(input("Enter a number to check if it's prime: "))
i=2
flag = True
while i<=num//2:
    if num % i == 0:
        flag = False
        break
    i += 1
if flag:
    print(f"{num} is a prime number.")
else:
    print(f"{num} is not a prime number.")
```

```python
#sum of the number 
num = int(input("Enter a positive integer: "))
sum = 0
while num > 0:
    sum += num %10
    num = num // 10
print(f"The sum of the digits is: {sum}")
```
